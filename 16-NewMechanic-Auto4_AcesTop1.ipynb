```python
# card Enum for playing cards
# shows the value/rank of the cards
# 2 is the weakest, Ace is the strongest
class Card(IntEnum):
    TWO = -6
    THREE = -5
    FOUR = -4
    FIVE = -3
    SIX = -2
    SEVEN = -1
    EIGHT = 0
    NINE = 1
    TEN = 2
    JACK = 3
    QUEEN = 4
    KING = 5
    ACE = 6

#Suit enum for playing cards
class Suit(Enum):
    SPADES = 'spades'
    CLUBS = 'clubs'
    HEARTS = 'hearts'
    DIAMONDS = 'diamonds'
```
