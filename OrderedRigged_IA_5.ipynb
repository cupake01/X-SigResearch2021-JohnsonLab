```python
from enum import Enum
from enum import IntEnum
from random import *
```

```python
# card Enum for playing cards
# shows the value/rank of the cards
# 2 is the weakest, Ace is the strongest
class Card(IntEnum):
    TWO = -6
    THREE = -5
    FOUR = -4
    FIVE = -3
    SIX = -2
    SEVEN = -1
    EIGHT = 0
    NINE = 1
    TEN = 2
    JACK = 3
    QUEEN = 4
    KING = 5
    ACE = 6
```

```python
#Suit enum for playing cards
class Suit(Enum):
    SPADES = 'spades'
    CLUBS = 'clubs'
    HEARTS = 'hearts'
    DIAMONDS = 'diamonds'
```

```python
# class to hold information for playing cards
# this is the actual playing card object(has a rank, suit)
class PlayingCard:
    def __init__(self, card_value, card_suit):
        self.card = card_value
        self.suit = card_suit
```

```python
full_deck = [] # this will be the original ordered deck
partial_deck = [] # to make sure you don't mess with the original ordered deck # keeps track of the remaining cards
player1_cards = [] # player1's hand
player2_cards = [] # player2's hand
```

```python
#creates a deck
#function to deal full deck of cards
# this only gives it to full deck
def create_deck():
    for card in Card :
        for suit in Suit:
            full_deck.append(PlayingCard(Card(card), Suit(suit)))
    return full_deck # return so the variable created earlier can u used(edit) for the entire. program
```
